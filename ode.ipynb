```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# example function
def f(x, t):
    # x: np.ndarray
    # t: float
    # k: float
    return - x
```

```python
x0 = np.array([2, 3])
t = np.linspace(0, 1, 10)
sol = odeint(f, x0, t)
t, sol
```

```python
plt.figure(figsize=(10, 7))
plt.plot(t, sol[:,0], label='x1')
plt.plot(t, sol[:,1], label='x2')
plt.legend();
```

```python
# example of f with additional arguments
def f_with_args(x, t, k, b):
    # x: np.ndarray
    # t: float
    # k: float
    return - k * x + b
```

```python
x0 = np.array([2, 3])
t = np.linspace(0, 1, 10)
k = 1
b = 4

sol = odeint(f_with_args, x0, t, args=(k, b))
t, sol
```
